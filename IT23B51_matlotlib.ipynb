```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.array([1,2,3,4])
y=np.array([4,6,8,9])
```

```python
x1=np.array([6,9,2,5])
y1=np.array([4,6,8,9])
```

```python
plt.scatter(x,y,color='r',s=20)
plt.scatter(x1,y1,s=40)
plt.show()
```

```python
plt.scatter(x,y,color='r',s=20)
plt.scatter(x,y,s=40,alpha=0.5)
plt.show()
```

```python
plt.scatter(x,y,color='r',s=20,marker="*")
plt.scatter(x,y,s=40,alpha=0.5)
plt.show()
```

```python
plt.scatter(x,y,color='r',s=20,marker="*")
plt.scatter(x1,y1,s=40,alpha=0.5)
plt.show()
```

```python
track_stud=np.array([20,30,60,10,80])
track_name=np.array(["web","data","mobile","cybersecurity","iot"])
plt.pie(track_stud)
plt.show()
```

```python
track_stud=np.array([20,30,60,10,80])
track_name=np.array(["web","data","mobile","cybersecurity","iot"])
plt.pie(track_stud,labels=track_name)
plt.show()
```

```python
track_stud=np.array([20,30,60,10,80])
track_name=np.array(["web","data","mobile","cybersecurity","iot"])
plt.pie(track_stud,labels=track_name,startangle=90)
plt.show()
```

```python
track_stud=np.array([20,30,60,10,80])
track_name=np.array(["web","data","mobile","cybersecurity","iot"])
ex=[0.4,0,0.3,0,0]
plt.pie(track_stud,labels=track_name,startangle=90,explode=ex)
plt.show()
```
